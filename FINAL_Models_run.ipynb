```python
def model_metrics(Y_test, y_pred, train_time, test_time, model_name):
    class_labels = Dataset['Label'].unique()
    print("======================================================================================================")
    print("MODEL : ", model_name)
    print("\n")
    encoded_values = dict(zip(le.classes_, range(len(le.classes_))))
    print(encoded_values)
    print("\n")
    # Compute the precision, recall, accuracy, and F1 score for each class
    for label in class_labels:
        precision = precision_score(Y_test, y_pred, average='weighted', labels=[label])
        recall = recall_score(Y_test, y_pred, average='weighted', labels=[label])
        accuracy = accuracy_score(Y_test, y_pred)
        f1 = f1_score(Y_test, y_pred, average='weighted', labels=[label])
        print(f'Class {label}: Accuracy={accuracy:.6f}, Precision={precision:.6f}, Recall={recall:.6f}, F1={f1:.6f}')

    # Compute the overall precision, recall, accuracy, and F1 score
    accuracy = accuracy_score(Y_test, y_pred)
    precision = precision_score(Y_test, y_pred, average='weighted')
    recall = recall_score(Y_test, y_pred, average='weighted')
    f1 = f1_score(Y_test, y_pred, average='weighted')


    print(f'Overall: Accuracy={accuracy:.6f}, Precision={precision:.6f}, Recall={recall:.6f}, F1={f1:.6f}')
    print(f"Time for {model_name} training is {train_time}")
    print(f"Time for {model_name} testing is {test_time}")
    print("======================================================================================================")
```

```python
def model_class_metrics (clf, X_train, Y_train, X_test, Y_test):
    import time
    import numpy as np
    from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

    # Initialize the dictionary to store the training and testing time for each class
    class_times = {}

    # Loop through each class
    for c in range(8):
        start_time = time.time()

        # Train the model on the current class
        clf.fit(X_train, (Y_train == c).astype(int))

        # Record the training time for the current class
        class_times['Class {}'.format(c)] = {'Training time': time.time() - start_time}

        start_time = time.time()

        # Test the model on the current class
        class_pred = clf.predict(X_test)

        # Record the testing time for the current class
        class_times['Class {}'.format(c)]['Testing time'] = time.time() - start_time

        # Calculate the metrics for the current class
        acc = accuracy_score((Y_test == c).astype(int), (class_pred == c).astype(int))
        cm = confusion_matrix((Y_test == c).astype(int), (class_pred == c).astype(int))
        cr = classification_report((Y_test == c).astype(int), (class_pred == c).astype(int))


        # Print the metrics for the current class
        print('Metrics for Class {}:'.format(c))
        print('Accuracy: {:.2f}'.format(acc))
        print('Confusion matrix:')
        print(cm)
        print('Classification report:')
        print(cr)

    # Print the training and testing time for each class
    print('Training and testing time for each class:')
    for c in class_times:
        print(c, class_times[c])
```
